```python
# Variables de acceso al fichero en la cuenta de almacenamiento de azure
account   = "profesor"
container = "tweets"
file_name = "tweets.txt"
sas       = "sp=r&st=2025-12-15T17:08:02Z&se=2025-12-30T01:23:02Z&spr=https&sv=2024-11-04&sr=c&sig=CKO9c6L8nPUlqJqbxpo%2FzXfjM8sIOx8qpQoGHwNMwNM%3D"
```

```python
# Configurar spark para que pueda acceder a contenedor de almacenamiento
spark.conf.set(f"fs.azure.sas.{container}.{account}.blob.core.windows.net", sas)
# Ruta al fichero
path = f"wasbs://{container}@{account}.blob.core.windows.net/{file_name}"
# Leer el fichero.
# Lo que deevuelve es un objeto de tipo Dataframe, distinto del RDD que nos generaba sparkCore.
# En este caso, se trata como si fuera una tabla EXCEL. Con filas y columnas
datos = spark.read.text(path)
display(datos)
```

```python
from pyspark.sql import functions as F 
# Renombrando columna 
datos = datos.withColumnRenamed("value", "tweet").withColumn("fecha", F.current_timestamp() )
display(datos)
```

```python
# Guardamos esa tabla de datos (los datos en bruto) dentro de Databricks, como una tabla de BBDD
spark.sql("CREATE DATABASE IF NOT EXISTS datosbruto")

datos.write.format("delta").mode("append").saveAsTable("datosbruto.tweets")
```
